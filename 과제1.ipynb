```python
import pandas as pd

pd.options.display.float_format = '{:.5f}'.format #e 나오는거 싫어서 바꿈

df = pd.read_excel('../data/05_car.csv')
print(df.head(n=10))
df = df.drop([0])
df_fillna = df.fillna(method='ffill') 
print(df_fillna.iloc[20:30, ])
a = df_fillna[df_fillna['시군구']=='계']
a
df_fillna_drop = df_fillna.drop([26,43,52,63,69,75,81,83,126,145,160,177,193,216,241,264,267])
df_fillna_drop.head()

#df_fillna_drop = df_fillna.drop([df_fillna['시군구']=='계'])
```

```python
seoul = df_fillna_drop.loc[df_fillna_drop['시도'] == '서울']
gg = df_fillna_drop.loc[df_fillna_drop['시도'] == '경기']
incheon = df_fillna_drop.loc[df_fillna_drop['시도'] == '인천']
seoul.head(n=10)
gg.head(n=10)
incheon.head(n=10)
```

```python
seoul_car1_mean = seoul['승용'].mean()
print("서울")
print(seoul_car1_mean)
print(seoul[seoul['승용'] > seoul['승용'].mean()].iloc[:, 0:3])
print("\n경기")
print(gg['승용'].mean())
print(gg[gg['승용'] > gg['승용'].mean()].iloc[:, 0:3])
print("\n인천")
print(incheon['승용'].mean())
print(incheon[incheon['승용'] > incheon['승용'].mean()].iloc[:, 0:3])
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import platform
import matplotlib

if platform.system() == 'Windows':
    matplotlib.rc('font', family='Malgun Gothic')
elif platform.system() == 'Darwin': # Mac
    matplotlib.rc('font', family='AppleGothic')
else: #linux
    matplotlib.rc('font', family='NanumGothic')

seoul_car1_mean_up = seoul[seoul['승용'] > seoul['승용'].mean()]
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(seoul_car1_mean_up['시군구'], seoul_car1_mean_up['승용'])
axes1.set_title('서울시 평균이상 구별 승용차 수')
axes1.set_xlabel('시군구')
axes1.set_ylabel('승용차 수')
```

```python
gg_car1_mean_up = gg[gg['승용'] > gg['승용'].mean()]
scatter_plot = plt.figure()
scatter_plot.tight_layout()
axes2 = scatter_plot.add_subplot(1,1,1)
axes2.scatter(gg_car1_mean_up['시군구'], gg_car1_mean_up['승용'])
axes2.set_title('경기도 평균이상 구별 승용차 수')
axes2.set_xlabel('시군구')
```

```python
incheon_car1_mean_up = incheon[incheon['승용'] > incheon['승용'].mean()]
scatter_plot = plt.figure()
axes3 = scatter_plot.add_subplot(1,1,1)
axes3.scatter(incheon_car1_mean_up['시군구'], incheon_car1_mean_up['승용'])
axes3.set_title('인천시 평균이상 구별 승용차 수')
axes3.set_xlabel('시군구')
axes3.set_ylabel('승용차 수')
```
